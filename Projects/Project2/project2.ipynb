# Section 5: data Science

The question that I am going to explore is **"Identify stocks that are potentially undervalued based on a combination of Price/Earnings, Earnings/Share, and Dividend Yield. Explore patterns among the top undervalued stocksâ€”are they concentrated in certain sectors?"**

##### The plan: 
1. Linear Regression: Fit a regression model using Dividend Yield and Earnings/Share as independent variables to predict Price. Stocks with actual prices much lower than predicted could be flagged as undervalued.

2. Hypothesis Testing: Use a t-test to check if the Dividend Yield or Earnings/Share of undervalued stocks is significantly different from other stocks.


#### Step 1: Linear Regression
We could set the filter for the price difference, for example, we select only those stocks where the price difference is greater than $50. This can help me to find the stocks that are significantly undervalued.


```python
from sklearn.linear_model import LinearRegression

x = df[['Earnings/Share', 'Dividend Yield']].dropna()
y = df.loc[x.index, 'Price']

model = LinearRegression()
model.fit(x, y)

df.loc[x.index, 'Predicted_Price'] = model.predict(x)

df['Price_Difference'] = df['Predicted_Price'] - df['Price']

undervalued_stocks = df[df['Price_Difference'] > 50]

undervalued_stocks_summary = undervalued_stocks[['Symbol', 'Name', 'Sector', 'Price', 'Predicted_Price', 'Price_Difference']]
undervalued_stocks_summary
```

#### Step 2: Hypothesis Test
We'll perform hypothesis testing to analyze if the Dividend Yield and Earnings/Share of these undervalued stocks differ significantly from other stocks. But frist, we are going find the T-Statistic and P-Value for Earnings/Share.

```python
from scipy.stats import ttest_1samp

population_mean_earnings = df['Earnings/Share'].mean()

earning = ttest_1samp(undervalued_stocks['Earnings/Share'], population_mean_earnings)
earning
```

Then, find the T-Statisticc and P-Value for Dividend Yield

```python
population_mean_dividend = df['Dividend Yield'].mean()

dividend = ttest_1samp(undervalued_stocks['Dividend Yield'], population_mean_earnings)
dividend
```
